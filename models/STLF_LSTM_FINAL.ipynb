```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')
```

```python


class EnergyDemandLSTM:
    def __init__(self, sequence_length=168, forecast_horizon=24):  # 1 week lookback, 1 day forecast
        self.sequence_length = sequence_length
        self.forecast_horizon = forecast_horizon
        self.target_scaler = MinMaxScaler(feature_range=(0, 1))
        self.feature_scaler = StandardScaler()
        self.model = None
        self.history = None

    def prepare_features(self, df):
        """Engineer additional time-based features"""
        df = df.copy()
        df['datetime'] = pd.to_datetime(df['datetime'])

        # Time-based features
        df['hour'] = df['datetime'].dt.hour
        df['day_of_week'] = df['datetime'].dt.dayofweek
        df['month'] = df['datetime'].dt.month
        df['quarter'] = df['datetime'].dt.quarter
        df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)

        # Cyclical encoding for time features
        df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
        df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
        df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
        df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
        df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
        df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

        # Lag features for demand
        for lag in [1, 24, 168]:  # 1 hour, 1 day, 1 week lags
            df[f'demand_lag_{lag}'] = df['nat_demand'].shift(lag)

        # Rolling statistics
        for window in [24, 168]:  # 1 day, 1 week windows
            df[f'demand_roll_mean_{window}'] = df['nat_demand'].rolling(window=window).mean()
            df[f'demand_roll_std_{window}'] = df['nat_demand'].rolling(window=window).std()

        return df

    def create_sequences(self, data, target_col='nat_demand'):
        """Create sequences for LSTM training"""
        # Select feature columns (exclude datetime and target)
        feature_cols = [col for col in data.columns if col not in ['datetime', target_col]]

        # Remove rows with NaN values
        data_clean = data.dropna()

        features = data_clean[feature_cols].values
        target = data_clean[target_col].values

        X, y = [], []

        for i in range(self.sequence_length, len(data_clean) - self.forecast_horizon + 1):
            # Input sequence
            X.append(features[i-self.sequence_length:i])
            # Output (next forecast_horizon steps)
            y.append(target[i:i+self.forecast_horizon])

        return np.array(X), np.array(y)

    def build_model(self, n_features):
        """Build improved LSTM architecture"""
        model = Sequential([
            # First LSTM layer with return sequences
            LSTM(128, return_sequences=True, input_shape=(self.sequence_length, n_features)),
            BatchNormalization(),
            Dropout(0.2),

            # Second LSTM layer
            LSTM(64, return_sequences=True),
            BatchNormalization(),
            Dropout(0.2),

            # Third LSTM layer
            LSTM(32, return_sequences=False),
            BatchNormalization(),
            Dropout(0.1),

            # Dense layers for output
            Dense(50, activation='relu'),
            Dropout(0.1),
            Dense(self.forecast_horizon, activation='linear')
        ])

        # Use custom optimizer with learning rate scheduling
        optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
        model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

        return model

    def fit(self, df, validation_split=0.2, epochs=100, batch_size=32, verbose=1):
        """Train the LSTM model"""
        print("Preparing features...")
        df_features = self.prepare_features(df)

        print("Creating sequences...")
        X, y = self.create_sequences(df_features)

        if len(X) == 0:
            raise ValueError("No valid sequences created. Check your data for sufficient non-NaN values.")

        # Scale features and target
        print("Scaling data...")
        X_reshaped = X.reshape(-1, X.shape[-1])
        X_scaled = self.feature_scaler.fit_transform(X_reshaped)
        X_scaled = X_scaled.reshape(X.shape)

        y_scaled = self.target_scaler.fit_transform(y.reshape(-1, 1)).reshape(y.shape)

        # Split data
        split_idx = int(len(X_scaled) * (1 - validation_split))
        X_train, X_val = X_scaled[:split_idx], X_scaled[split_idx:]
        y_train, y_val = y_scaled[:split_idx], y_scaled[split_idx:]

        print(f"Training data shape: {X_train.shape}")
        print(f"Validation data shape: {X_val.shape}")

        # Build model
        self.model = self.build_model(X.shape[-1])

        if verbose:
            self.model.summary()

        # Define callbacks
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=8, min_lr=1e-7, verbose=verbose),
            ModelCheckpoint('best_lstm_model.h5', monitor='val_loss', save_best_only=True, verbose=0)
        ]

        # Train model
        print("Training model...")
        self.history = self.model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=callbacks,
            verbose=verbose
        )

        return self

    def predict(self, df, start_idx=None):
        """Make predictions"""
        df_features = self.prepare_features(df)

        if start_idx is None:
            start_idx = len(df_features) - self.sequence_length

        # Get the last sequence for prediction
        feature_cols = [col for col in df_features.columns if col not in ['datetime', 'nat_demand']]
        last_sequence = df_features[feature_cols].iloc[start_idx:start_idx + self.sequence_length].values

        # Scale the sequence
        last_sequence_scaled = self.feature_scaler.transform(last_sequence)
        last_sequence_scaled = last_sequence_scaled.reshape(1, self.sequence_length, -1)

        # Predict
        prediction_scaled = self.model.predict(last_sequence_scaled, verbose=0)

        # Inverse transform
        prediction = self.target_scaler.inverse_transform(prediction_scaled)

        return prediction.flatten()

    def evaluate(self, df_test):
        """Evaluate model performance"""
        df_features = self.prepare_features(df_test)
        X_test, y_test = self.create_sequences(df_features)

        if len(X_test) == 0:
            print("No test sequences available")
            return {}

        # Scale test data
        X_test_reshaped = X_test.reshape(-1, X_test.shape[-1])
        X_test_scaled = self.feature_scaler.transform(X_test_reshaped)
        X_test_scaled = X_test_scaled.reshape(X_test.shape)

        # Predict
        y_pred_scaled = self.model.predict(X_test_scaled, verbose=0)
        y_pred = self.target_scaler.inverse_transform(y_pred_scaled.reshape(-1, 1)).reshape(y_pred_scaled.shape)

        # Calculate metrics
        mse = mean_squared_error(y_test.flatten(), y_pred.flatten())
        mae = mean_absolute_error(y_test.flatten(), y_pred.flatten())
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test.flatten(), y_pred.flatten())

        # Mean Absolute Percentage Error
        mape = np.mean(np.abs((y_test.flatten() - y_pred.flatten()) / y_test.flatten())) * 100

        metrics = {
            'MSE': mse,
            'RMSE': rmse,
            'MAE': mae,
            'R2': r2,
            'MAPE': mape
        }

        return metrics, y_test, y_pred

    def plot_training_history(self):
        """Plot training history"""
        if self.history is None:
            print("No training history available")
            return

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

        # Loss plot
        ax1.plot(self.history.history['loss'], label='Training Loss')
        ax1.plot(self.history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()

        # MAE plot
        ax2.plot(self.history.history['mae'], label='Training MAE')
        ax2.plot(self.history.history['val_mae'], label='Validation MAE')
        ax2.set_title('Model MAE')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('MAE')
        ax2.legend()

        plt.tight_layout()
        plt.show()

# Example usage and demonstration
def main():
    # Load and prepare data
    print("Loading data...")
    try:
        dataset = pd.read_csv("continuous dataset.csv")
    except FileNotFoundError:
       print("File not found. Please check the file path.")
       return

    print("Data loaded successfully.
    print(f"Dataset shape: {dataset.shape}")
    print(f"Date range: {dataset['datetime'].min()} to {dataset['datetime'].max()}")

    # Split data into train and test
    dataset['datetime'] = pd.to_datetime(dataset['datetime'])
    dataset = dataset.sort_values('datetime')

    # Use 80% for training, 20% for testing
    split_idx = int(len(dataset) * 0.8)
    train_data = dataset.iloc[:split_idx].copy()
    test_data = dataset.iloc[split_idx:].copy()

    print(f"Training data: {len(train_data)} samples")
    print(f"Test data: {len(test_data)} samples")

    # Initialize and train model
    lstm_model = EnergyDemandLSTM(sequence_length=168, forecast_horizon=24)

    try:
        lstm_model.fit(train_data, validation_split=0.2, epochs=50, batch_size=32, verbose=1)

        # Plot training history
        lstm_model.plot_training_history()

        # Evaluate on test data
        print("\nEvaluating model...")
        metrics, y_true, y_pred = lstm_model.evaluate(test_data)

        print("\nModel Performance:")
        for metric, value in metrics.items():
            print(f"{metric}: {value:.4f}")

        # Plot predictions vs actual
        fig, ax = plt.subplots(figsize=(15, 8))

        # Plot first few predictions for clarity
        n_samples_to_plot = min(5, len(y_true))

        for i in range(n_samples_to_plot):
            time_idx = np.arange(i * 24, (i + 1) * 24)
            ax.plot(time_idx, y_true[i], 'b-', alpha=0.7, label='Actual' if i == 0 else '')
            ax.plot(time_idx, y_pred[i], 'r--', alpha=0.7, label='Predicted' if i == 0 else '')

        ax.set_xlabel('Hours')
        ax.set_ylabel('Energy Demand')
        ax.set_title('LSTM Predictions vs Actual Energy Demand')
        ax.legend()
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

        # Make a single prediction
        print("\nMaking a 24-hour forecast...")
        forecast = lstm_model.predict(test_data, start_idx=-200)  # Use last available data

        print("24-hour forecast:")
        for i, pred in enumerate(forecast):
            print(f"Hour {i+1}: {pred:.2f}")

    except Exception as e:
        print(f"Error during training or evaluation: {e}")
        print("This might be due to insufficient data or TensorFlow installation issues.")

if __name__ == "__main__":
    main()
```
